```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score,f1_score
import seaborn as sns
```

```python
print('24MAI0111_ANMOL_CHAUBEY')
#read dataset
df = pd.read_csv('decision_tree_social_network_ads.csv')
```

```python
df.head()
```

```python
#feature and target selection 
x = df.iloc[:,[2,3]].values
y = df.iloc[:,4].values
```

```python
#splitting data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.25, random_state=42)
```

```python
#standardise the features by scaling
sstand_scal_x = StandardScaler()
x_train = sstand_scal_x.fit_transform(x_train)
x_test = sstand_scal_x.transform(x_test)
```

```python
#Initialize the Decision Tree Classifier with entropy criterion
classifier = DecisionTreeClassifier(criterion='entropy', random_state=42)
```

```python
#train the Decision Tree Classifier on the training data
classifier.fit(x_train, y_train)
```

```python
#predict the target values for the testing set
y_pred = classifier.predict(x_test)
```

```python
#calculate the performance matrics
print('24MAI0111_ANMOL_CHAUBEY')
conf_matrix = confusion_matrix(y_test,y_pred)
accu = accuracy_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
print("Confusion Matrix :")
print(conf_matrix)
print(f"Accuracy : {accu:.2f}")
print(f"Precision: {prec:.2f}")
print(f"Recall : {recall:.2f}")
print(f"F1 Score:{f1:.2f}")
```

```python
#plot and display confusion matrix 
plt.figure(figsize=(4,3))
sns.heatmap(conf_matrix,annot=True,fmt='d',cmap='Blues',xticklabels =['Not Purchased','Purchased'],yticklabels=['Not Purchased','Purchased'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix by 24MAI0111')
plt.show()
```

```python
#plot and display the decision tree visualization.
plt.figure(figsize=(12,8))
plot_tree(classifier, filled=True, feature_names=['Age','EstimatedSalary'])
plt.title('Decision Tree Visualization by 24MAI0111')
plt.show()
```
